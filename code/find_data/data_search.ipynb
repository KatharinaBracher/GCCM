## CauseEffectPairs (Mooij)

Information for pair0043:

Mean Daily Air temperature near surface (.995 sigma level) on a 144x73 grid (2.5 degree) of day 50 and day 51 of year 2000.

Each data pair consists of 144 × 73 − 143 = 10369 data points, distributed across the globe. 143 data points were subtracted because at the north pole values are repeated across all longitudes.

x: year 2000, day 50
y: year 2000, day 51

units: K

Data source:
http://www.esrl.noaa.gov/psd/data/gridded/data.ncep.reanalysis.surface.html


ground truth:
x --> y
